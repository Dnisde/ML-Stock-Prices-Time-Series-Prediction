```python
%pylab inline
import numpy
from sklearn.decomposition import KernelPCA
from sklearn.metrics import explained_variance_score
from sklearn.datasets import load_iris # select dataset
from sklearn.datasets import load_diabetes
iris = load_iris()
diabates = load_diabetes()

X = iris.data # feature
y = iris.target # label

```

```python
y
```

```python
X.shape
```

```python
diabates
```

```python
# input
figure()
title("Nonlinear input")
scatter(X[y==0, 0], X[y==0, 1], color='red', marker='^', alpha=0.5)
scatter(X[y==1, 0], X[y==1, 1], color='blue', marker='o', alpha=0.5)
scatter(X[y==2, 0], X[y==2, 1], color='green', marker='*', alpha=0.5)
xlabel('X')
ylabel('y')

# Kernel PCA
kpca = KernelPCA(n_components=2, kernel="rbf") # set component number and kernel
X_kpca = kpca.fit_transform(X)
```

```python
figure()
title("KernelPCA with rbf")
scatter(X_kpca[y == 0, 0], X_kpca[y == 0, 1], color='red', marker='^', alpha=0.5)
scatter(X_kpca[y == 1, 0], X_kpca[y == 1, 1], color='blue', marker='o', alpha=0.5)
scatter(X_kpca[y == 2, 0], X_kpca[y == 2, 1], color='green', marker='*', alpha=0.5)
xlabel('PC1')
ylabel('PC2')

show()
```
